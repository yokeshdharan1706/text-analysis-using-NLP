```python
import requests as req
from bs4 import BeautifulSoup
import pandas as pd 

def extracting(url):
    class_lis=["td-post-content tagdiv-type","tdb-block-inner td-fix-index"] 
    response = req.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    text_content=[]
    for i in class_lis:
        my_file = soup.find_all('div', class_=i)
        for element in my_file:
            text_content.append(element.get_text(strip=True))
    return text_content
```

```python
def file_write(my_id, text_content):
    file_path = f'C:/Users/HP/Desktop/jupyter projects/extracted content/{my_id} .txt'
    with open(file_path, 'w+',encoding='utf-8') as file:
        for item in text_content:
            file.write(item)
            file.write("\n")
```

```python
df = pd.read_excel('weblink.xlsx')
my_id = df['URL_ID'].tolist()
my_link = df['URL'].tolist()
for i, url in enumerate(my_link):
    text_content = extracting(url)
    if text_content:
        print(f"{my_id[i]}")
        print(text_content)
        print("Writing to file...")
        file_write(str(my_id[i]), text_content)
        print("Write complete.")
        print()
```
